### Notebook Formatting & Style Helpers

### Tesla Stock Price v.s. Search Volume

**Challenge:** Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes. Label one axis 'TSLA Stock Price' and the other 'Search Trend'. 

```python
plt.figure(figsize=(10,6))
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(df_tesla['MONTH'], df_tesla['TSLA_USD_CLOSE'], color='#F11515')
ax2.plot(df_tesla['MONTH'], df_tesla['TSLA_WEB_SEARCH'], color='b')

ax1.set_ylabel('TSLA Stock Price', color='#F11515')
ax2.set_ylabel('Search Trend', color='b')
plt.show()
```

```python
# increases size and resolution
plt.figure(figsize=(14,8), dpi=120)
# Increases the size and rotate the labels on the x-axis
plt.xticks(fontsize=14, rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(df_tesla['MONTH'], df_tesla['TSLA_USD_CLOSE'], color='#F11515', linewidth=3)
ax2.plot(df_tesla['MONTH'], df_tesla['TSLA_WEB_SEARCH'], color='b', linewidth=3)

ax1.set_ylabel('TSLA Stock Price', color='#F11515', fontsize=14)
ax2.set_ylabel('Search Trend', color='b', fontsize=14)

# set Set the minimun and maximun values on the axes
ax1.set_xlim([df_tesla['MONTH'].min(), df_tesla['MONTH'].max()])
ax1.set_ylim([0, 600])

# format the ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

plt.title('Tesla Web Search vs Price', fontsize=18)

#Displays the chart explicity
plt.show()
```

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate.

```python
plt.figure(figsize=(14,8), dpi=120)
plt.title('Unemployment Benefits in the US vs the U/E Rate', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.set_ylabel('FREUD U/E Rate', fontsize=14, color='#5B2C6F')
ax2.set_ylabel('Search Trend', fontsize=14, color='skyblue')

ax1.set_xlim([df_unemployment['MONTH'].min(), df_unemployment['MONTH'].max()])
ax1.set_ylim([0, 12])

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

#plot
ax1.plot(df_unemployment['MONTH'], df_unemployment['UNRATE'],  color='#5B2C6F', linestyle='dashed')
ax2.plot(df_unemployment['MONTH'], df_unemployment['UE_BENEFITS_WEB_SEARCH'], color='skyblue')
ax1.grid(color='grey', linestyle='--')
plt.show()
```

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
roll_unemployment = df_unemployment.rolling(window=6, on='MONTH').mean()
#df_unemployment.info()
```

```python
plt.figure(figsize=(14,8), dpi=120)
plt.title('Rolling Monthly US "Unemployment Benefits" Web Searches vs the UNRATE', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.set_ylabel('FREUD U/E Rate', fontsize=14, color='#5B2C6F')
ax2.set_ylabel('Search Trend', fontsize=14, color='skyblue')

ax1.set_xlim([df_unemployment['MONTH'][0], df_unemployment['MONTH'].max()])
ax1.set_ylim([3, 10.5])

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

#plot
ax1.plot(df_unemployment['MONTH'], roll_unemployment['UNRATE'],  color='#5B2C6F', linestyle='dashed')
ax2.plot(df_unemployment['MONTH'], roll_unemployment['UE_BENEFITS_WEB_SEARCH'], color='skyblue')
ax1.grid(color='grey', linestyle='--')
plt.show()
```
